```python
import google.generativeai as genai
from kaggle_secrets import UserSecretsClient
import PIL.Image
from IPython.display import display
```

```python
user_secrets = UserSecretsClient()
secret_value_0 = user_secrets.get_secret("KEY")
genai.configure(api_key=secret_value_0)
```

```python
model=genai.GenerativeModel('gemini-1.5-flash-latest')
response=model.generate_content('Merhaba')
print(response.text)
```

```python
response=model.generate_content('Bana bir şaka anlat')
print("Şaka:", response.text)
```

```python
response=model.generate_content('Basit bir sayı oyunu kodu yaz.')
print(response.text)
```
